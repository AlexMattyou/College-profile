```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generating example data
np.random.seed(0)
X = np.random.rand(100, 1)  # Independent variable
y = 2 + 3 * X + np.random.randn(100, 1)  # Dependent variable with true intercept=2 and slope=3
```

```python
# Fitting models
model_no_bias = LinearRegression(fit_intercept=False)
model_no_bias.fit(X, y) 
y_pred_no_bias = model_no_bias.predict(X)

model_with_bias = LinearRegression(fit_intercept=True)
model_with_bias.fit(X, y)
y_pred_with_bias = model_with_bias.predict(X)
```

```python
# Plotting comparison
plt.figure(figsize=(12, 6))
plt.scatter(X, y, label='Data points', alpha=0.7)
plt.plot(X, y_pred_no_bias, color='red', linewidth=2, label='Regression without bias')
plt.plot(X, y_pred_with_bias, color='blue', linewidth=2, label='Regression with bias')
plt.legend()
plt.title('Comparing Regression Models: With vs Without Bias Term')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.grid(True)
plt.show()
```

```python
# Calculating performance metrics
mse_no_bias = mean_squared_error(y, y_pred_no_bias)
r2_no_bias = r2_score(y, y_pred_no_bias)

mse_with_bias = mean_squared_error(y, y_pred_with_bias)
r2_with_bias = r2_score(y, y_pred_with_bias)

# Displaying model parameters and performance
print("\n=== Model Without Bias (fit_intercept=False) ===")
print(f"Slope: {model_no_bias.coef_[0][0]:.4f}")
print(f"MSE: {mse_no_bias:.4f}")
print(f"R-squared: {r2_no_bias:.4f}")

print("\n=== Model With Bias (fit_intercept=True) ===")
print(f"Intercept: {model_with_bias.intercept_[0]:.4f}")
print(f"Slope: {model_with_bias.coef_[0][0]:.4f}")
print(f"MSE: {mse_with_bias:.4f}")
print(f"R-squared: {r2_with_bias:.4f}")
```

```python
# Residual plot comparison
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(y_pred_no_bias, y - y_pred_no_bias, color='red', alpha=0.7)
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Residuals Without Bias')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')

plt.subplot(1, 2, 2)
plt.scatter(y_pred_with_bias, y - y_pred_with_bias, color='blue', alpha=0.7)
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Residuals With Bias')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')

plt.tight_layout()
plt.show()
```
