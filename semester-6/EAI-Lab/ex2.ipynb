```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generating example data
np.random.seed(0)
X = np.random.rand(100, 1)  # Independent variable
y = 2 + 3 * X + np.random.randn(100, 1)  # Dependent variable with noise

# Creating a DataFrame for easier analysis
data = pd.DataFrame(data=np.hstack([X, y]), columns=['X', 'y'])

# Exploratory Data Analysis (EDA)
print("=== Basic Data Exploration ===")
print(data.head())  # Show first few rows
print("\nData Statistics:")
print(data.describe())  # Show basic statistics
```

```python
# Scatter plot to visualize relationship
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['y'], alpha=0.7)
plt.title('Scatter plot of X vs y (Before Regression)')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.grid(True)
plt.show()
```

```python
# Calculating correlation coefficient
correlation = data['X'].corr(data['y'])
print(f'\nCorrelation coefficient between X and y: {correlation:.4f}')
```

```python
# Fitting a linear regression model
model = LinearRegression()
model.fit(X, y)

# Getting model parameters
intercept = model.intercept_[0]
slope = model.coef_[0][0]
r_squared = model.score(X, y)

print('\n=== Regression Model Results ===')
print(f'Intercept (b0): {intercept:.4f}')
print(f'Slope (b1): {slope:.4f}')
print(f'R-squared: {r_squared:.4f}')
```

```python
# Plotting the regression line with scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['y'], alpha=0.7, label='Actual Data')
plt.plot(data['X'], model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression Model Fit')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Residual analysis (important part of EDA for regression)
residuals = y - model.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(model.predict(X), residuals, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```
